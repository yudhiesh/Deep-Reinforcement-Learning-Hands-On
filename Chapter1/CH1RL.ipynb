# Reinforcement Learning Complications 

1. RL depends on an agent's behavioor and, to some extent, it is the result of this behvaior. If your agent decided to do inefficient things, then the observations will tell you nothing about what it has done wrong and what should be done to improve the outcome(agent gives a negative feedback all the time). If the agent is subborn and keeps making mistakes, then the observations will give the false impression that there is no way to get a larger reward - life is suffering - which could be completely wrong. This can be rephrased as having non-i.i.d. data. 

2. The agent has to not only exploit the knowledge that it has learned but to actively explore the environment as doing things differently might significantly improve the outcome. Too much exploration might also decrease the reward, so a balance needs to be found. 

3. Rewards can be delayed, such as in chess where a particular move might not show the benefits until much later on in the game.

### Reward

Reward is a scalar value we obtain periodically from the environment. The purpose of the reward is to inform the agent how well it is doing in the environment and this reward can be a positive, negative, large, small but it is a number nonetheless. 

We don't define how frequently the agent receives the reward but it is common practice to receive rewards every fixed timestamp or at every environment interaction, just for convenience. 

In the case of once-in-a-lifetiome reward systems, all rewards except the last ones will be zero.

Reward is local meaning that it reflects the success of the agent's recent activity and not all the successes achieved byu the agent so far. The agent is trying to achieve the largest accumulated reward over its sequence of actions. 

1. Financial trading: An amount of profit is a reward for a trader buying and selling stocks.

2. Chess: Reward is obtained at the end of the game as a win, lose, or draw. Of course, it's up to interpretation. For me, for example, achieving a
draw in a match against a chess grandmaster would be a huge reward. In practice, we need to specify the exact reward value, but it could be a fairly complicated expression. For instance, in the case of chess, the reward could be proportional to the opponent's strength.

3. Dopamine system in the brain: There is a part of the brain (limbic system) that produces dopamine every time it needs to send a positive signal to the rest of the brain. Higher concentrations of dopamine lead to a sense
of pleasure, which reinforces activities considered by this system to be good. Unfortunately, the limbic system is ancient in terms of the things it considers good—food, reproduction, and dominance—but that is a totally different story!

4. Computer games: They usually give obvious feedback to the player, which is either the number of enemies killed or a score gathered. Note in this example that reward is already accumulated, so the RL reward for arcade games should be the derivative of the score, that is, +1 every time a new enemy
is killed and 0 at all other time steps.

5. Web navigation: There are problems, with high practical value, that require the automated extraction of information available on the web. Search engines are trying to solve this task in general, but sometimes, to get to the data you're looking for, you need to fill in some forms or navigate through a series of links, or complete CAPTCHAs, which can be difficult for search engines
to do. There is an RL-based approach to those tasks in which the reward is the information or the outcome that you need to get.

6. Neural network (NN) architecture search: RL has been successfully applied to the domain of NN architecture optimization, where the aim is to get the best performance metric on some dataset by tweaking the number of layers or their parameters, adding extra bypass connections, or making other changes to the NN architecture. The reward in this case is the performance (accuracy or another measure showing how accurate the NN predictions are).

7. Dog training: If you have ever tried to train a dog, you know that you need to give it something tasty (but not too much) every time it does the thing you've asked. It's also common to punish your pet a bit (negative reward) when it doesn't follow your orders, although recent studies have shown that this isn't as effective as a positive reward.

8. School marks: We all have experience here! School marks are a reward system designed to give pupils feedback about their studying.

### The environment 

The environment is everything outside of an agent. In the most general sense, it's the rest of the universe, but this goes slightly and exceeds the capacity of even tomorrow's computers. 

### Actions

Actions are things that an agent can do in the environment. Actins can, for example, be moves allowed by the rules of play. They can be as simple as move pawn one space forward or as complicatede as fill the tax from in for tomorrow morning. 

### Observations

Observations of the environment form the second information channel for an agent with the first being the reward. Observations are pieces of information that the source provides the agent about what is going on in the environment. Where they differ compared to the environment is that the environment contains everything that the agent is exposed compared to the observations which is part of the environment but a very small amount which is related to maxiziming the goal. 

1. Financial trading: Here, the environment is the whole financial market and everything that influences it. This is a huge list of things, such as the latest news, economic and political conditions, weather, food supplies, and Twitter trends. Even your decision to stay home today can potentially indirectly influence the world's financial system (if you believe in the "butterfly effect"). However, our observations are limited to stock prices, news, and so on. We don't have access to most of the environment's state, which makes trading such a nontrivial thing.

2. Chess: The environment here is your board plus your opponent, which includes their chess skills, mood, brain state, chosen tactics, and so on. Observations are what you see (your current chess position), but, at some levels of play, knowledge of psychology and the ability to read an opponent's mood could increase your chances.

3. Dopamine system: The environment here is your brain plus your nervous system and your organs' states plus the whole world you can perceive. Observations are the inner brain state and signals coming from your senses.

4. Computer game: Here, the environment is your computer's state, including all memory and disk data. For networked games, you need to include other computers plus all Internet infrastructure between them and your machine. Observations are a screen's pixels and sound only. These pixels are not a tiny amount of information (somebody calculated that the total number of possible moderate-size images (1024×768) is significantly larger than the number of atoms in our galaxy), but the whole environment state is definitely larger.

5. Web navigation: The environment here is the Internet, including all the network infrastructure between the computer on which our agent works and the web server, which is a really huge system that includes millions and millions of different components. The observation is normally the web page that is loaded at the current navigation step.

6. NN architecture search: In this example, the environment is fairly simple and includes the NN toolkit that performs the particular NN evaluation and the dataset that is used to obtain the performance metric. In comparison to the Internet, this looks like a tiny toy environment.Observations might be different and include some information about testing, such as loss convergence dynamics or other metrics obtained from the evaluation step.

7. Dog training: Here, the environment is your dog (including its hardly observable inner reactions, mood, and life experiences) and everything around it, including other dogs and even a cat hiding in a bush. Observations are signals from your senses and memory.

8. School: The environment here is the school itself, the education system of the country, society, and the cultural legacy. Observations are the same as for the dog training example—the student's senses and memory.

## Theoretical foundations of RL

### The Markov process or the Markov chain 

> ***The future is independent of the past given the present***

Imagine that you have some system in front of you that you can only observe. What you observe is called states, and the system can switch between these states according to the laws of dynamics. 

All possible states for a system form a set called the **state space** with a finite value. 

Your observations form a sequence of states or a **chain**. 

Simple example using the weather observed on a particular day. 

We can observe the day to either be rainy or sunny which is the **state space**. 

A sequence of observations over time forms a chain of states such as [sunny, rainy, sunny, sunny, rainy] and this is called the **history**.

To call the system a MP it needs to fulfill the Markov property, which means that the future system dynamics from any state have to depend on this state only. The main point of the Markov property is to make every observable state self-contained to desceribe the future of the system. 

In the weather example, the Markov property limits our model to represent only the cases when a sunny day can be followed by a raining day with the same probability, regardless of the amount of sunny days we've seen in the past. 

If you wanted to make the model more complex then you could add in more dependencies such as the season, in which the state space will look like [sunny+summer, sunny+winter, rainy+summer, rainy+winter].

If the system complies with the Markov property then you can represent the transition probabilities in a transition matrix.

If we have a sunny day, then there is an 80% chance that the next day will be sunnny and a 20% chance that the next day will be rainy.

The formal definition of the MP is:

1. A set of states (**S**) that a system can be in

2. A transition matrix(**T**), with transition probabilities, which defines the system dynamics.

We assume that our office worker's weekday usually starts from the **Home** state and that he starts his day with **Coffee** without exception.

In practice we rarely have the luxury of knowing the exact transition matrix. Usually we only have observations of our current systems states and this is called **episodes**. 

It's also worth noting that the Markov property implies stationarity where the underlying transition distribution for any state does not change over time.

### Markov reward processes

Reward can be represented by another square matrix similar to the transition matrix where a reward given for transitioning from state *i* to *j* will reside in the row *i* and column *j*. A value 𝛾 is added in which is a single number in the range of 0 to 1(inclusive). This number represents the discount factor and it stands for the foresightedness of the agent. If gamma equals 1, then return *G*, just equals a sum of all subsequent rewards and corresponds to the agent that has perfect visibility of any subsequent rewards. If gamma equals to 0, *G* will be jsut immeidate reward without any subsequent state and will correspond to absolute short-sightednesss. The gamma parameter is important in RL and think about it as a measure of how far into the future we look to estimate the future return.

We observe a chain of state transitions in an MP which is still the same for a Markov reward process but for every transition, we have ourextra quantity - *reward*.

For every episode, we define **return** at the time, *t*, as this quantity:

\begin{equation}
G_{t}=R_{t+1}+\gamma R_{t+2}+\cdots=\sum_{k=0}^{\infty} \gamma^{k} R_{t+k+1}
\end{equation}

This return quantity is not very useful in practivce, as it was defined for very specific chain we observed from our Markov reward process, so it can vary widely, even for the same state. But youu can calculate the mathematical expectation of return fior any state (by averaging a large number of chains), you will get the **value of the state**:

\begin{equation}
V(s)=\mathbb{E}\left[G \mid S_{t}=s\right]
\end{equation}

This interpretation is simple - for every state, *s*, the value , *V(s)*, is the average(or expected) return we get by following the Markov reward process.

We will start with a simple case: gamma = 0.

How do you calculate the values of states here? To answer this question, let's fix our state to Chat. What could the subsequent transition be? The answer is that it depends on chance. According to our transition matrix for the Dilbert process, there is a 50% probability that the next state will be Chat again, 20% that it will be Coffee, and 30% that it
will be Computer. When gamma = 0, our return is equal only to a value of the next immediate state. 

So, if we want to calculate the value of the Chat state, then we need to sum all transition values and multiply that by their probabilities:

V(chat) = –1 * 0.5 + 2 * 0.3 + 1 * 0.2 = 0.3

V(coffee) = 2 * 0.7 + 1 * 0.1 + 3 * 0.2 = 2.1

V(home) = 1 * 0.6 + 1 * 0.4 = 1.0

V(computer) = 5 * 0.5 + (–3) * 0.1 + 1 * 0.2 + 2 * 0.2 = 2.8

So, **Computer** is the most valuable state to be in if you care about immediate reward. 


Now what would the value be when gamma = 1? The answer is that the value is infinite for all state. Our diagram doesn't contain *sink* states (states without outgoing transitions), and when our discount equals to 1, we care about a potentially infinite number of transitions in the future. When the gamma is equal to 0, all the values are positive in the short term, so the sum of the infinite number of positive values will give us an inifinte value, regardless of the starting state. 

This infinite results shows us one of the reasons to introduce gamma into a Markov reward process instead of just summing all future rewards. In most cases the process can have a very large or an infinite number of transitions, as it is not practical to deal with infinite values, we would like to limit the horizon at which we calculate values for. Gamma values of less than 1 provide that limitation for us.

### Policy 
The simple defintion of a policy is that it is some set of rules that controls the agent's behavior. Even fort a fairly simple environements, we can have a variety of policies. With a robot as an example we could do many things such as:

1. Blindly move forward regardless of anything

2. Try to go around obstacles by checking whether that previous forward action failed. 

With different policies it gives us different returns which makes it important to find a good policy. The policy is defined as the probablity distribution over actions for every possible state:

\begin{equation}
\pi(a \mid s)=P\left[A_{t}=a \mid S_{t}=s\right]
\end{equation}

If the policy is fixed and not changing then the  MDP becomes a Markov reward process, as we can reduce the transition and reward matrices with a policy's probabilitites and get rid of the action dimensions.
